Question 6: Plot GameStop Stock Graph

Use the make_graph function to graph the GameStop Stock Data, also provide a title for the graph.

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# Step 1: Get GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)

# Step 2: Define or reuse make_graph function
def make_graph(stock_data, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'], name="Close Price"))
    fig.update_layout(title=title, xaxis_title='Date', yaxis_title='Stock Price (USD)')
    fig.show()

# Step 3: Plot GameStop stock graph
make_graph(gme_data, 'GameStop Stock Price Over Time')
```
